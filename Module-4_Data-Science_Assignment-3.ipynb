```python
# Question 1. Scenario: A company wants to analyze the sales performance of its products in different regions. They have collected the following data:
'''   
    Region A: [10, 15, 12, 8, 14]
    Region B: [18, 20, 16, 22, 25]
    Calculate the mean sales for each region.

'''



# Solution: To calculate the mean sales for each region, we need to add up all the values in the data set and divide this number by the number of values.
'''
For Region A, the mean sales is calculated as follows: (10 + 15 + 12 + 8 + 14) / 5 = 11.8

For Region B, the mean sales is calculated as follows: (18 + 20 + 16 + 22 + 25) / 5 = 20.2

So, the mean sales for Region A is 11.8 and for Region B is 20.2.

'''


# Code: Here is the solution code for the asked question:

region_a = [10, 15, 12, 8, 14]
region_b = [18, 20, 16, 22, 25]

mean_sales_region_a = sum(region_a) / len(region_a)
mean_sales_region_b = sum(region_b) / len(region_b)

print("Mean sales for Region A:", mean_sales_region_a)
print("Mean sales for Region B:", mean_sales_region_b)
```

```python
# Question 5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
'''
    Group A: [85, 90, 92, 88, 91]
    Group B: [82, 88, 90, 86, 87]
    Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.

'''


# Solution :
'''
Here are the important steps, we have to follow to Perform a t-test to determine if there is a significant difference in the mean scores between the two groups:

1. Calculate the mean score for each group:
   - Group A: (85 + 90 + 92 + 88 + 91) / 5 = 89.2
   - Group B: (82 + 88 + 90 + 86 + 87) / 5 = 86.6

2. Calculate the standard deviation for each group:
   - Group A: sqrt(((85 - 89.2)^2 + (90 - 89.2)^2 + (92 - 89.2)^2 + (88 - 89.2)^2 + (91 - 89.2)^2) / (5 - 1)) = 2.87
   - Group B: sqrt(((82 - 86.6)^2 + (88 - 86.6)^2 + (90 - 86.6)^2 + (86 - 86.6)^2 + (87 - 86.6)^2) / (5 - 1)) = 2.42

3. Calculate the pooled standard deviation:
   - sp = sqrt(((n1-1)s1^2 + (n2-1)s2^2) / (n1+n2-2))
   - sp = sqrt(((5-1) * 8.25 + (5-1) * 5.84) / (5+5-2)) = sqrt(68.16 / 8) = **3**

4. Calculate the t-value:
   - t = (x̄1 - x̄2) / (s_p * sqrt(1/n1 + 1/n2))
   - t = (89.2 - 86.6) / (3 * sqrt(1/5 + 1/5)) = **1.14**

5. Determine the degrees of freedom:
   - df = n1 + n2 - 2 = **8**

6. Look up the critical value of t for your desired level of significance and degrees of freedom:
   - For a two-tailed test with alpha = .05 and df = 8, the critical value of t is **±2.306**.

7. Compare your calculated t-value to the critical value of t:
   - Since |t| < |t_critical|, we fail to reject the null hypothesis that there is no significant difference in the mean scores between the two groups.

Therefore, based on this analysis, we can conclude that there is no significant difference in the mean scores between the two groups.

'''


# Code: Here is the solution code for the asked question:

import scipy.stats as stats

group_a = [85, 90, 92, 88, 91]
group_b = [82, 88, 90, 86, 87]

t_statistic, p_value = stats.ttest_ind(group_a, group_b)

print("The t-statistic is:", t_statistic)
print("The p-value is:", p_value)
```

```python
# Question 8. Scenario: A company wants to analyze the relationship between employee tenure and job satisfaction. The data collected is as follows:
'''
    Employee Tenure (in years): [2, 3, 5, 4, 6, 2, 4]
    Job Satisfaction (on a scale of 1 to 10): [7, 8, 6, 9, 5, 7, 6]
    Perform a linear regression analysis to predict job satisfaction based on employee tenure.

'''


# Solution :
'''
Here’s how we can perform a linear regression analysis to predict job satisfaction based on employee tenure:

Plot the data:

You can use a scatter plot to visualize the relationship between employee tenure and job satisfaction.
Calculate the correlation coefficient between employee tenure and job satisfaction:

You can use the same formula we used earlier to calculate the correlation coefficient between advertising expenditure and sales.
Calculate the slope and intercept of the regression line:

Slope = Covariance / Variance of Employee Tenure
Intercept = Mean of Job Satisfaction - Slope * Mean of Employee Tenure
Use the regression line to predict job satisfaction for a given employee tenure:

Job Satisfaction = Intercept + Slope * Employee Tenure

'''


# Code: Here is the solution code for the asked question:

import numpy as np
import matplotlib.pyplot as plt

employee_tenure = [2, 3, 5, 4, 6, 2, 4]
job_satisfaction = [7, 8, 6, 9, 5, 7, 6]

correlation_coefficient = np.corrcoef(employee_tenure, job_satisfaction)[0][1]

slope = correlation_coefficient * (np.std(job_satisfaction) / np.std(employee_tenure))

intercept = np.mean(job_satisfaction) - slope * np.mean(employee_tenure)

print("Slope:", slope)
print("Intercept:", intercept)

plt.scatter(employee_tenure, job_satisfaction)

x = np.linspace(0, 7)
y = intercept + slope * x
plt.plot(x, y)

plt.xlabel("Employee Tenure (in years)")
plt.ylabel("Job Satisfaction (on a scale of 1 to 10)")

plt.show()
```

```python
# Question 11. Scenario: A quality control department wants to test the weight consistency of a product. The weights of a sample of products are as follows:
'''
    [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]
    Perform a hypothesis test to determine if the mean weight differs significantly from 10 grams

'''

# Solution : Here's how we can perform a hypothesis test to determine if the mean weight differs significantly from 10 grams:
'''
1. State the null and alternative hypotheses:
   - Null hypothesis (H0): The mean weight is equal to 10 grams.
   - Alternative hypothesis (Ha): The mean weight is not equal to 10 grams.

2. Set the significance level:
   - Let's assume a significance level of 0.05.

3. Calculate the sample mean and standard deviation:
   - Sample mean = (10.2 + 9.8 + 10.0 + 10.5 + 10.3 + 10.1) / 6 = 10.2
   - Sample standard deviation = sqrt(((10.2 - 10)^2 + (9.8 - 10)^2 + (10.0 - 10)^2 + (10.5 - 10)^2 + (10.3 - 10)^2 + (10.1 - 10)^2) / (6 - 1)) = 0.214

4. Calculate the t-statistic:
   - t = (sample mean - hypothesized mean) / (sample standard deviation / sqrt(sample size))
   - t = (10.2 - 10) / (0.214 / sqrt(6))
   - t = 1.512

5. Determine the degrees of freedom:
   - Degrees of freedom = sample size - 1 = 6 - 1 = 5

6. Determine the critical value of t for a given significance level and degrees of freedom:
   - You can use a t-distribution table or an online calculator to determine the critical value of t.

7. Compare the t-statistic to the critical value of t:
   - If the absolute value of the t-statistic is greater than the critical value of t, then reject the null hypothesis and conclude that the mean weight differs significantly from 10 grams.

In this case, the critical value of t for a two-tailed test with a significance level of 0.05 and degrees of freedom of 5 is approximately ±2.571. Since the absolute value of the calculated t-statistic is less than the critical value of t, we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest that the mean weight differs significantly from 10 grams.

'''

# Code: Here is the solution code for the asked question:

import math
import scipy.stats as stats

data = [10.2, 9.8, 10.0, 10.5, 10.3, 10.1]
hypothesized_mean = 10
significance_level = 0.05

sample_mean = sum(data) / len(data)
sample_standard_deviation = math.sqrt(sum([(x - sample_mean) ** 2 for x in data]) / (len(data) - 1))

t_statistic, p_value = stats.ttest_1samp(data, hypothesized_mean)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

if abs(t_statistic) > stats.t.ppf(1 - significance_level / 2, len(data) - 1):
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# Question 12. Scenario: A company wants to analyze the click-through rates of two different website designs. The number of clicks for each design is as follows:
'''
    Design A: [100, 120, 110, 90, 95]
    Design B: [80, 85, 90, 95, 100]
    Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.

'''

# Solution :

# Code: Here is the solution code for the asked question:

import numpy as np
import scipy.stats as stats

design_a_clicks = [100, 120, 110, 90, 95]
design_b_clicks = [80, 85, 90, 95, 100]

observed_frequencies = np.array([design_a_clicks, design_b_clicks])

expected_frequency = np.sum(observed_frequencies) / observed_frequencies.size

expected_frequencies = np.full(observed_frequencies.shape, expected_frequency)

chi_square_statistic, p_value, degrees_of_freedom, expected_frequencies = stats.chi2_contingency(observed_frequencies)

print("Chi-Square Statistic:", chi_square_statistic)
print("P-Value:", p_value)
print("Degrees of Freedom:", degrees_of_freedom)
print("Expected Frequencies:\n", expected_frequencies)

if chi_square_statistic > 9.488:
    print("There is a significant difference in the click-through rates between Design A and Design B.")
else:
    print("There is no significant difference in the click-through rates between Design A and Design B.")
```

```python
# Question 17. Scenario: A study is conducted to compare the performance of three different machine learning algorithms. The accuracy scores for each algorithm are as follows:
'''
    Algorithm A: [0.85, 0.80, 0.82, 0.87, 0.83]
    Algorithm B: [0.78, 0.82, 0.84, 0.80, 0.79]
    Algorithm C: [0.90, 0.88, 0.89, 0.86, 0.87]
    Perform a Kruskal-Wallis test to determine if there is a significant difference in the median accuracy scores between the algorithms.

'''
# Solution :
'''
The Kruskal-Wallis test is a non-parametric test used to determine whether there is a statistically significant difference between the medians of three or more independent groups. It is the non-parametric equivalent of the one-way ANOVA and is typically used when the normality assumption is violated.
The null hypothesis of the Kruskal-Wallis test is that all groups have the same median value. Since the p-value is greater than 0.05 (the typical significance level), we fail to reject the null hypothesis and conclude that there is not enough evidence to suggest that there is a significant difference in the median accuracy scores between the algorithms.

'''

# Code: Here is the solution code for the asked question:

from scipy.stats import kruskal

a = [0.85, 0.80, 0.82, 0.87, 0.83]
b = [0.78, 0.82, 0.84, 0.80, 0.79]
c = [0.90, 0.88, 0.89, 0.86, 0.87]

statistic, p_value = kruskal(a, b, c)

print("Kruskal-Wallis H-test")
print("Statistic:", statistic)
print("p-value:", p_value)
```

```python
# Question 18. Scenario: A company wants to analyze the effect of price on sales. The data collected is as follows:
'''
    Price (in dollars): [10, 15, 12, 8, 14]
    Sales: [100, 80, 90, 110, 95]
    Perform a simple linear regression to predict sales based on price.

'''

# Solution :

# Code: Here is the solution code for the asked question:

import numpy as np
from scipy.stats import linregress

price = [10, 15, 12, 8, 14]
sales = [100, 80, 90, 110, 95]

slope, intercept, r_value, p_value, std_err = linregress(price, sales)

print("Slope:", slope)
print("Intercept:", intercept)
print("R-squared:", r_value**2)

# Predict sales for a price of $13
price_new = 13
sales_new = slope * price_new + intercept
print("Predicted sales for a price of $13:", sales_new)
```

```python
# Question 19. Scenario: A survey is conducted to measure the satisfaction levels of customers with a new product. The data collected is as follows:
'''
    [7, 8, 9, 6, 8, 7, 9, 7, 8, 7]
    Calculate the standard error of the mean satisfaction score.

'''

# Solution : To calculate the standard error of the mean satisfaction score, we can use the following formula:
'''
    standard error = standard deviation / sqrt(n)

where n is the sample size.
'''

# Code: Here is the solution code for the asked question:

import numpy as np

data = [7, 8, 9, 6, 8, 7, 9, 7, 8, 7]
n = len(data)

std_dev = np.std(data)
std_err = std_dev / np.sqrt(n)

print("Standard error of the mean satisfaction score:", std_err)
```

```python
# Question 20. Scenario: A company wants to analyze the relationship between advertising expenditure and sales. The data collected is as follows:
'''
    Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]
    Sales (in thousands): [25, 30, 28, 20, 26]
    Perform a multiple regression analysis to predict sales based on advertising expenditure.

'''

# Solution :

# Code: Here is the solution code for the asked question:

import numpy as np
import statsmodels.api as sm

advertising = [10, 15, 12, 8, 14]
sales = [25, 30, 28, 20, 26]

X = np.array(advertising).reshape(-1, 1)
y = np.array(sales)

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())
```
